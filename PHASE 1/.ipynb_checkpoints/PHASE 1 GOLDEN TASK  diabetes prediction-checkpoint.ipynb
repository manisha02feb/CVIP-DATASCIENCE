```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("diabetes_prediction_dataset.csv")
```

```python
data.head()
```

```python
#Checking duplicates (duplicates in data can cause problem or can help you in certain ways , but to not complicate our problem we will remove it)
duplicates = data.duplicated().sum()
print(duplicates)
```

```python
#This will remove all the duplicates in the dataset
data = data.drop_duplicates()
```

```python
data.info()
```

```python
unique_values = data.nunique()
print(unique_values)
```

```python
data.isna().sum()
#to find out the missing columns values
```

```python
data.gender.value_counts()
```

```python
data.drop(data[data['gender']=='Other'].index , inplace=True)
#removing other as gender as their count is very low 
```

```python
data.gender.value_counts()
```

```python
data.smoking_history.value_counts()
```

```python
data["gender"].value_counts().plot(kind="bar", color =["pink", "black"],edgecolor=["pink","black"],figsize=(10,6))
plt.title("Gender Distribution")

plt.ylabel("Count");
#male vs female in the dataset
```

```python
sns.countplot(x='HbA1c_level',data=data)
```

```python
plt.hist(data["age"],bins=25,color="red", edgecolor="black")
plt.title("Age Distribution")
plt.xlabel("Age(in years)")
plt.ylabel("counts")
plt.show()
# age distribution in our dataset
```

```python
sns.countplot(x='age',data=data)
```

```python
sns.boxplot(x='smoking_history',y='blood_glucose_level',data=data)
```

```python
sns.histplot(data["bmi"],kde=True,bins=30,color="red")
plt.title("Bmi Distribution");
#visualising bmi 
```

```python
plt.hist(data["smoking_history"],bins=20);
#visualing smoking history
```

```python
data["diabetes"].value_counts().plot(kind="bar",color=["blue","red"],figsize=(8,5));
#diabetes comparison


```

```python
data["hypertension"].value_counts().plot(kind="bar",color=["blue","red"],figsize=(8,5));
#hypertension comparison
```

```python
sns.boxplot(data= data, x=data["diabetes"], y=data["age"]); # most diabetic patients between 50-80
#visualising age vs diabetes
```

```python
sns.boxplot(data=data , x=data["diabetes"], y = data["bmi"]) 
#visualising bmi vs diabetes
```

From this graph we can conclude that most patients that have diabetes have BMI between 28-40 ( does not mean that patients with higher BMI cannot have diabetes , this graph just represents for most of the patients , not all of them)

```python
sns.boxplot(data = data, x=data["diabetes"], y=data["HbA1c_level"])
#visualising hb1ac level vs diabetes 
```

From this boxplot we can conclude that most of the patients that have diabetes have HbA1c_level between 6-8. ( does not mean that patients with higher HbA1c_level cannot have diabetes , this graph just represents for most of the patients , not all of them )

```python
sns.scatterplot(data = data, x=data["age"],y=data["bmi"],hue=data["diabetes"])
# visuallising a scatter plot between age and bmi
```

From this graph we can conclude that if the patient age is above 40 and BMI is above 28 , then the chances of diabetes increase drastically . 


```python
sns.scatterplot(data = data , x=data["blood_glucose_level"],y=data["HbA1c_level"], hue=data["diabetes"])
#visuallising a scatter plot between blooad glucose level and hba1c level
```

```python
sns.scatterplot(data = data, x=data["age"],y=data["blood_glucose_level"],hue=data["diabetes"]);
#visualising scatter plot between age and blood glucose level 
```

We can without a doubt conclude from this graph that if your blood_glucose_level is above 200 , no matter what the patients age is there is 99-100% probability that the patient has diabetes

One more thing that we can conclude from this graph is that if the patients age is above 50 and has blood_glucose_level between 125-170 there is a chance that the patient might have diabetes

```python
sns.scatterplot(data = data, x=data["age"],y=data["HbA1c_level"],hue=data["diabetes"]);
#visualising a scatter plot between age and hbaic level
```

We can conclude from this plot that if a patient has over 6.7 HbA1c_level, whatever be the age it is 98-100% probability that a patient has diabetes

Also we can conclude that if a patient has a HbA1c_level above 5.8 and age above 40 chances increase of the patient being diagnosed with diabetes
